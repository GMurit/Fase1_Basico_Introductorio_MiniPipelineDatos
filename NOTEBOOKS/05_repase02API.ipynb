```python
import requests
import pandas as pd
```

```python
#URL DE LA API QUE VAMOS A USAR
url = "https://api.coingecko.com/api/v3/coins/markets"

params = {
    "vs_currency": "usd",
    "order": "market_cap_desc",
    "per_page": 10,
    "page": 1,
    "sparkline": False
}

#HACEMOS LA PETICIÓN A LA API

response = requests.get(url, params = params)

response.status_code
```

```python
#TRANSFORMAMOS LOS DATOS RECIBIDOS EN UN DICCIONARIO DE PYTHON
data = response.json()
```

```python
#AHORA CONVERTIMOS EL DICCIONARIO EN UN DATAFRAME
df = pd.DataFrame(data)
df
df.head()
```

```python
#VAMOS A CREAR UNA COLUMNA EURO PARA SABER A QUE PRECIO ESTAN
df["precio_en_euros"] = df["current_price"] * 0.95
df
```

```python
df_mayores = df[df["market_cap"]>10_000_000]
df_mayores
```

```python
resumen = df_mayores.groupby(df_mayores["symbol"].str[0])["current_price"].mean().reset_index()
resumen.rename(columns={"current_price":"precio_medio"}, inplace = True)
resumen
```
